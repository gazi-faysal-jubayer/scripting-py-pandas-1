```python
female = []
male = []
child = []
shopkeeper = []
elder = []
for i in range(21):
    x = ((female_df['exit time(s)'] > i*10) & (female_df['exit time(s)'] <= (i+1)*10)).sum()
    y = ((male_df['exit time(s)'] > i*10) & (male_df['exit time(s)'] <= (i+1)*10)).sum()
    z = ((child_df['exit time(s)'] > i*10) & (child_df['exit time(s)'] <= (i+1)*10)).sum()
    a = ((shopkeeper_df['exit time(s)'] > i*10) & (shopkeeper_df['exit time(s)'] <= (i+1)*10)).sum()
    b = ((elder_df['exit time(s)'] > i*10) & (elder_df['exit time(s)'] <= (i+1)*10)).sum()
    female.append(x)
    male.append(y)
    child.append(z)
    shopkeeper.append(a)
    elder.append(b)
```

```python
df1['Female'] = female
df1['Male'] = male
df1['Child'] = child
df1['Shopkeeper'] = shopkeeper
df1['Elder'] = elder
```
